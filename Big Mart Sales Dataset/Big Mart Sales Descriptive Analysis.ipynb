### Measures of Central Tendency for 'Item_MRP' column

```python
# mean
df['Item_MRP'].mean()

# median
df['Item_MRP'].median()

# mode
df['Item_MRP'].mode()
```

### Measures of Spread for 'Item_MRP' column

```python
# Range
np.ptp(df['Item_MRP'])

# Variance
df['Item_MRP'].var()

# Standard Deviation
df['Item_MRP'].std()
```

### Measures of Shape for 'Item_Outlet_Sales' column

```python
# Skewness
sts.skew(df['Item_Outlet_Sales'])

# Kurtosis
sts.kurtosis(df['Item_Outlet_Sales'], fisher=True)
sts.kurtosis(df['Item_Outlet_Sales'], fisher=False)
```

### Detect Outliers on 'Item_Outlet_Sales' columns using IQR

```python
def outlier_finder(column):
    Q3 = np.quantile(df[column], 0.75)
    Q1 = np.quantile(df[column], 0.25)
    
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q3 + 1.5*IQR
    
    return df[ (df[column] < lower_limit) | (df[column] > upper_limit) ]
```

```python
outlier_finder('Item_Outlet_Sales')
```

### Detect & Remove Outliers from 'Item_MRP' & 'Item_Outlet_Sales' columns using z-score

```python
z = np.abs(sts.zscore(df[['Item_MRP', 'Item_Outlet_Sales']]))
z
```

```python
np.where(z > 3)
df.shape
```

```python
# In order to remove the outliers
df_cleaned = df[(z < 3).all(axis=1)] # axis=1: for rows
df_cleaned
```
